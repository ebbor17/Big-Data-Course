```python
import pandas as pd
df_swing = pd. read_csv('2008_swing_states.csv')
df_swing[['state', 'county','total_votes','dem_votes', 'dem_share']]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
_= plt.hist(df_swing['dem_share'])
_= plt.xlabel('percent of votes for Obama')
_= plt.ylabel('Number of Counties')
plt.show
```

```python
_= sns.swarmplot(x='state', y='dem_share', data=df_swing)
_= plt.xlabel('state')
_= plt.ylabel('percent of vates for Obama')
plt.show
```

```python
import numpy as np
x = np.sort(df_swing['dem_share'])
y = np.arange(1, len(x)+1) / len(x)
_= plt.plot(x, y, marker='.', linestyle='none')
_= plt.xlabel('Percent of votes for Obama')
_= plt.ylabel('Empirical Cumulative Distribution Function')
_= plt.margins(0.01)
plt.show
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df_swing = pd. read_csv('2008_swing_states.csv')
df_swing[['state', 'county','total_votes','dem_votes', 'dem_share']]
from sklearn.linear_model import LinearRegression
x = df_swing['dem_votes'].values.reshape(-1,1)
y =  df_swing['total_votes'].values.reshape(-1,1)
lr = LinearRegression()
lr.fit(x,y)
Y_pred = lr.predict(x)

plt.scatter(x,y)
plt.plot(x,Y_pred, color='red')
plt.show
```
